```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500 ):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
msp_onehot = pd.get_dummies(msp_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
msp_onehot['Neighborhood'] = msp_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [msp_onehot.columns[-1]] + list(msp_onehot.columns[:-1])
msp_onehot = msp_onehot[fixed_columns]

msp_onehot.head()
```

```python
msp_grouped = msp_onehot.groupby('Neighborhood').mean().reset_index()
msp_grouped
```

```python
num_top_venues = 5

for hood in msp_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = msp_grouped[msp_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]


num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
city_venues_sorted = pd.DataFrame(columns=columns)
city_venues_sorted['Neighborhood'] = msp_grouped['Neighborhood']

for ind in np.arange(msp_grouped.shape[0]):
    city_venues_sorted.iloc[ind, 1:] = return_most_common_venues(msp_grouped.iloc[ind, :], num_top_venues)

city_venues_sorted.head()
```

```python
kclusters = 10

msp_grouped_clustering = msp_grouped.drop('Neighborhood', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(msp_grouped_clustering)

kmeans.labels_[0:10] 
```

```python
city_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
msp_merged = sort_result
msp_merged = msp_merged.join(city_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
msp_merged
```

```python
relocation_data  = pd.DataFrame(columns = ['Current Neighborhood', 'Current City','Target Neighborhood', 'Target City'])
n=0
for i in range(data_m.shape[0]):
    for j in range(data_sp.shape[0]):
        if (data_m['Cluster Labels'][i]==data_sp['Cluster Labels'][j] and data_m['Cluster Labels Price&Incr'][i]==data_sp['Cluster Labels Price&Incr'][j]):
            relocation_data.loc[n,'Current Neighborhood']=data_sp['Neighborhood'][j]
            relocation_data.loc[n,'Current City']=data_sp['City'][j]
            relocation_data.loc[n,'Target Neighborhood']=data_m['Neighborhood'][i]
            relocation_data.loc[n,'Target City']=data_m['City'][i]
            n=n+1
        else:
            continue
```

```python
relocation_data.groupby(['Current Neighborhood','Current City','Target City'])['Target Neighborhood'].apply(', '.join).reset_index()
```

```python
relocation_data_suggested_v = pd.DataFrame(columns = ['Current Neighborhood', 'Current City','Suggested Neighborhood_Venue','Target City'])
n=0
for i in range(data_m.shape[0]):
    for j in range(data_sp.shape[0]):
        if (data_m['Cluster Labels'][i]==data_sp['Cluster Labels'][j]):
            relocation_data_suggested_v.loc[n,'Current Neighborhood']=data_sp['Neighborhood'][j]
            relocation_data_suggested_v.loc[n,'Current City']=data_sp['City'][j]
            relocation_data_suggested_v.loc[n,'Suggested Neighborhood_Venue']=data_m['Neighborhood'][i]
            relocation_data_suggested_v.loc[n,'Target City']=data_m['City'][i]
            n=n+1
        else:
            continue
```

```python
vdata = relocation_data_suggested_v.groupby(['Current Neighborhood','Current City','Target City'])['Suggested Neighborhood_Venue'].apply(', '.join).reset_index()
pd.set_option('display.max_colwidth',300)
vdata 
```

```python
relocation_data_suggested_p = pd.DataFrame(columns = ['Current Neighborhood', 'Current City','Suggested Neighborhood_Price','Target City'])
n=0
for i in range(data_m.shape[0]):
    for j in range(data_sp.shape[0]):
        if (data_m['Cluster Labels Price&Incr'][i]==data_sp['Cluster Labels Price&Incr'][j]):
            relocation_data_suggested_p.loc[n,'Current Neighborhood']=data_sp['Neighborhood'][j]
            relocation_data_suggested_p.loc[n,'Current City']=data_sp['City'][j]
            relocation_data_suggested_p.loc[n,'Suggested Neighborhood_Price']=data_m['Neighborhood'][i]
            relocation_data_suggested_p.loc[n,'Target City']=data_m['City'][i]
            n=n+1
        else:
            continue
```
